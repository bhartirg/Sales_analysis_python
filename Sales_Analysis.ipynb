# Exploratory Data Analysis

### Analysis Based on Gender

```python
# plotting a bar chart for Gender and it's count

ax = sns.countplot(x = 'Gender',data = df,hue="Gender",dodge=False)

for bars in ax.containers:
    ax.bar_label(bars)
sns.set(rc={'figure.figsize':(5,6)})
```

```python
# plotting a bar chart for gender vs total amount

sales_gen = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
plt.title('Gender vs Amount')
x=sns.barplot(x = 'Gender',y= 'Amount' ,hue='Gender', data = sales_gen ,errorbar=None,palette={'F': 'salmon', 'M': 'skyblue'},dodge=False)
sns.set(rc={'figure.figsize':(5,6)})
#for bars in x.containers:
 #   x.bar_label(bars)
```

*Based on the graphical representations provided, we can see that most of the buyers are females and even the purchasing power of females are greater than men*

### Analysis Based on Age

```python
ax = sns.countplot(data = df, x = 'Age Group', hue = 'Gender')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# Total Amount vs Age Group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)
custom_palette = sns.color_palette("tab10", 7)

sns.barplot(x = 'Age Group',y= 'Amount',hue='Amount',data = sales_age,palette=custom_palette , dodge=False)
sns.set(rc={'figure.figsize':(15,5)})
```

*Based on the graphical representations provided, we can see that most of the buyers are of age group between 26-35 yrs female*

### Analysis Based on State

```python
# total number of orders from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(15,5)})
color_palette = sns.color_palette("Blues_r", len(sales_state))
#base_color = "#2ca02c"  # Green color

# Create a custom color palette using a dark to light green color scheme
#custom_palette = sns.light_palette(base_color, n_colors=10).as_hex()
sns.barplot(data = sales_state, x = 'State',y= 'Orders',hue='State', palette=color_palette)

```

```python
# total amount/sales from top 10 states

sales_state = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)
#base_color = "#2ca02c"  # Green color

# Create a custom color palette using a dark to light green color scheme
#custom_palette = sns.light_palette(base_color, n_colors=10).as_hex()
sns.set(rc={'figure.figsize':(15,5)})
sns.barplot(data = sales_state, x = 'State',y= 'Amount',hue='State')
```

*Based on the graphical representations provided, we can see that most of the orders & total sales/amount are from Uttar Pradesh, Maharashtra and Karnataka respectively*


### Analysis Based on Marital Status

```python
ax = sns.countplot(data = df, x = 'Marital_Status')

sns.set(rc={'figure.figsize':(5,7)})
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(6,5)})
sns.barplot(data = sales_state, x = 'Marital_Status',y= 'Amount', hue='Gender')
```

*Based on the graphical representations provided, we can see that most of the buyers are married (women) and they have high purchasing power*

### Analysis Based on Occupation

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Occupation',hue='Occupation')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Occupation',y= 'Amount',hue='Occupation')
```

*Based on the graphical representations provided, we can see that most of the buyers are working in IT, Healthcare and Aviation sector.*

### Analysis Based on Product Category

```python
sns.set(rc={'figure.figsize':(20,5)})
ax = sns.countplot(data = df, x = 'Product_Category',hue='Product_Category')

for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_state = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_Category',y= 'Amount',hue='Product_Category')
```

*Based on the graphical representations provided, we can see that most of the sold products are from Food, Clothing and Electronics category.*

```python
sales_state = df.groupby(['Product_ID'], as_index=False)['Orders'].sum().sort_values(by='Orders', ascending=False).head(10)

sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(data = sales_state, x = 'Product_ID',y= 'Orders',hue='Product_ID')
```
